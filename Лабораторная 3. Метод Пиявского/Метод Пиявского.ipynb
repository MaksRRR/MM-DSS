```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
a = -20
b = -5
f = lambda x: np.cos(x)+np.cos((2/7)*x)

Xs = np.arange(a, b, 0.001)

plt.plot(Xs, f(Xs))
plt.show()
```

```python
len(Xs)
```

```python
%%time


def estimate_L(h):
    return 2 * np.max(abs(f(Xs[:-1]) - f(Xs[1:]))/h)

def g_func(u):
    return f(u) - L*(abs(Xs-u))


h = (Xs[1:] - Xs[:-1])[0]
L = estimate_L(h)

u = []
p = []
u.append(a)
u.append(b)
p.append(g_func(u[0]))
p.append(np.maximum(p[-1], g_func(u[1])))

k = 0
itrs = 1000
eps = 1e-5
while (k < itrs+1) & (abs(u[-1]-u[-2]) > eps):
    k+=1
    u.append(Xs[np.argmin(p[-1])])
    p.append(np.maximum(p[-1], g_func(u[-1])))
```

```python
k
```

```python
for i in range(2, k, 75):
    plt.subplots(figsize=(10, 5), dpi=100)
    plt.plot(Xs, f(Xs), color='black')
    plt.plot(Xs, p[i], color='green')
    plt.scatter(u[i], f(u[i]), color='red', zorder=2)
    plt.legend()
plt.show()
```
