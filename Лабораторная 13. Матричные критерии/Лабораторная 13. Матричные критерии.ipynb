```python
import numpy as np
```

```python
N = 3
matrix = np.array([
  [30+N, 10, 20, 25+N/2],
  [50, 70-N, 10+N/2, 25],
  [25-N/2, 35, 40, 60-N/2]
])

p = np.array([0.3, 0.2, 0.4, 0.1])
```

```python
def bayes(matrix, p):
    criteria_vec = matrix @ p
    print('Оптимальная стратегия по критерию Байеса:', np.argmax(criteria_vec)+1)
    return tuple(criteria_vec), np.argmax(criteria_vec)
```

```python
bayes_res = bayes(matrix, p)
```

```python
bayes_res
```

```python
def laplas(matrix):
    p = [1 / matrix.shape[1]] * matrix.shape[1]
    criteria_vec = matrix @ p
    print('Оптимальная стратегия по критерию Лапласа:', np.argmax(criteria_vec)+1)
    return tuple(criteria_vec), np.argmax(criteria_vec)
```

```python
laplas_res = laplas(matrix)
```

```python
def vald(matrix):
    criteria_vec = matrix.min(axis=1)
    print('Оптимальная стратегия по критерию Вальда:', np.argmax(criteria_vec)+1)
    return tuple(criteria_vec), np.argmax(criteria_vec)
```

```python
vald_res = vald(matrix)
```

```python
def savage(matrix):
    criteria_vec = (matrix.max(axis=0) - matrix).max(axis=1)
    print('Оптимальная стратегия по критерию Сэвиджа:', np.argmin(criteria_vec)+1)    
    return tuple(criteria_vec), np.argmin(criteria_vec)
```

```python
savage_res = savage(matrix)
```

```python
def hurwitz(matrix, const):
    criteria_vec = const * matrix.min(axis=1) + (1 - const) * matrix.max(axis=1)
    print('Оптимальная стратегия по критерию Гурвица:', np.argmax(criteria_vec)+1)
    return tuple(criteria_vec), np.argmax(criteria_vec)
```

```python
hurwitz(matrix, 0.1)
```

В зависимости от факторов среды и внутренней структуры можно вырать наиболее подходящую оценку.
